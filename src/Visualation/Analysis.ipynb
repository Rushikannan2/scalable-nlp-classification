```python
import matplotlib.pyplot as plt
import numpy as np

models = [
    "Normal Logistic Regression",
    "Balanced Logistic Regression",
    "Naive Bayes",
    "Support Vector Machine (SVM)",
    "FastText",
    "LSTM",
    "BiLSTM"
]
test_acc = [90.11, 88, 85, 91.21, 87, 80.20, 79.86]

# Custom color palette
colors = [
    "#4CAF50",
    "#2196F3",
    "#FFC107",
    "#F44336",
    "#9C27B0",
    "#FF9800",
    "#607D8B"
]

# Find best model
best_idx = np.argmax(test_acc)

# Highlight best model
colors[best_idx] = "#000000"

plt.figure(figsize=(11, 6))

bars = plt.barh(models, test_acc, color=colors, edgecolor='black')

plt.xlabel("Test Accuracy (%)", fontsize=12)
plt.title("Model Performance Comparison (Test Accuracy)", fontsize=14, fontweight='bold')

plt.xlim(75, 95)

# Add values
for i, bar in enumerate(bars):
    width = bar.get_width()
    label = f"{width:.2f}%"
    
    if i == best_idx:
        label += "  ⭐"
    
    plt.text(width + 0.3,
             bar.get_y() + bar.get_height()/2,
             label,
             va='center',
             fontsize=10,
             fontweight='bold' if i == best_idx else 'normal')

plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Models
models = [
    "Normal Logistic Regression",
    "Balanced Logistic Regression",
    "Naive Bayes",
    "SVM",
    "FastText",
    "LSTM",
    "BiLSTM"
]

# Accuracies
train_acc = [
    92.52,
    None,
    None,
    97.45,
    90.87,
    84.22,
    84.96
]

val_acc = [
    90.43,
    88,
    85,
    91.41,
    87.28,
    79.83,
    79.61
]

# Convert to numpy (for better handling)
train_arr = np.array([np.nan if x is None else x for x in train_acc])
val_arr = np.array(val_acc)

x = np.arange(len(models))

# Create plot
plt.figure(figsize=(10, 6))

# Plot lines
plt.plot(x, train_arr, marker='o', linestyle='-', linewidth=2, label="Train Accuracy")
plt.plot(x, val_arr, marker='s', linestyle='--', linewidth=2, label="Validation Accuracy")

# Highlight best validation model
best_idx = np.argmax(val_arr)
plt.scatter(best_idx, val_arr[best_idx], s=150, edgecolors='black', zorder=5)
plt.text(best_idx, val_arr[best_idx] + 0.5,
         f"Best: {models[best_idx]}",
         ha='center', fontsize=9, fontweight='bold')

# Annotations
for i, (t, v) in enumerate(zip(train_arr, val_arr)):
    if not np.isnan(t):
        plt.text(i, t + 0.3, f"{t:.1f}", ha='center', fontsize=8)
    plt.text(i, v - 1.2, f"{v:.1f}", ha='center', fontsize=8)

# Labels & title
plt.xticks(x, models, rotation=30)
plt.xlabel("Models", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.title("Model Performance Comparison (Train vs Validation)", fontsize=14, fontweight='bold')

# Grid + legend
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.tight_layout()
plt.show()
```
